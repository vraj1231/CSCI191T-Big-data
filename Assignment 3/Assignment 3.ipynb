```python
import pandas as pd
import matplotlib.pyplot as plt 

data = pd.read_csv("movies_metadata.csv")      #reading the data from pandas 

data1 = pd.DataFrame(data, columns = ['budget'])    # only taking into consideration of budget column as instructed
data1.head()  #checking the column
```

```python
data["budget"] =pd.to_numeric(data['budget'],errors='coerce') #using to numeric function to convert strings into float 

print ("Mean of budget",data["budget"].mean()) # getting the mean of the column
```

```python
print("Variance of budget",data["budget"].var())  #getting the variance of the budget
```

```python
new_data = data1.query("budget != \'0\'")  #using query to eliminate all zeroes in the column budget

print( "mean of cleaned budget",pd.to_numeric(new_data["budget"], errors = 'coerce').mean())  #mean without zeroes
print( "variance of cleaned budget",pd.to_numeric(new_data["budget"], errors = 'coerce').var()) #variance without zerores

new_data.head()  #outputting using head function
```

```python
data3 = pd.read_csv("ratings.csv")   #reading the file rating from kaggle
data3.head()
```

```python
data3["rating"].describe()  #using describe function to see the column types
```

```python
import seaborn as sns # using the seaborn library to plot the countplot graph

map1 = data3["rating"]   #using the column
sns.set(color_codes=True) #plotting the graph
sns.distplot(map1) #outputting the graph
```

```python
sns.countplot(x='rating', data=data3) #bargraph of the column rating
```

```python
dict1 = {'0.5':0, '1.0':0,'1.5':0,'2.0':0,'2.5':0,'3.0':0,'3.5':0,'4.0':0,'4.5':0,'5.0':0} #using dict to get number of count
for x in data3["rating"]:                                                                  # of each item in rating column
    if x == 0.5:
        dict1['0.5'] += 1
    elif x ==  1.0:
        dict1['1.0'] += 1
    elif x ==  1.5:
        dict1['1.5'] += 1
    elif x ==  2.0:
        dict1['2.0'] += 1
    elif x ==  2.5:
        dict1['2.5'] += 1
    elif x ==  3.0:
        dict1['3.0'] += 1
    elif x == 3.5:
        dict1['3.5'] += 1
    elif x == 4.0:
        dict1['4.0'] += 1
    elif x == 4.5:
        dict1['4.5'] += 1
    elif x == 5.0:
        dict1['5.0'] += 1
            
print(dict1)
```

```python
total = sum(dict1.values())   #dispersion of the all the rating items                 
for k in dict1:
     
    print (k, ":", '%.2f' %((dict1[k] / total) * 100))
```

```python
plt.pie(dict1.values(), labels=dict1.keys(), startangle=90, autopct='%.1f%%',radius=2.3) #piechart of column rating
plt.show()
```
